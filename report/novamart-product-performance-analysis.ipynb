```python

# importing library
import pandas as pd

# Did not wanna override current dataframe and created a copy to keep data integrity
nans_solved = products_cleaned_df.copy()

# To handle missing values, I decided to fill the missing values in with the category median price

nans_solved['price'] = nans_solved.groupby('category')['price'].transform(lambda x: x.fillna(x.median()))
nans_solved['price'] = nans_solved['price'].fillna(nans_solved['price'].median())

# I created a column that shows wheter a columns Nans were imputed or not.   True - Imputed Nan --- False NoNulls
nans_solved['Nans filled'] = products_cleaned_df['price'].isna()

nans_solved.head(20)
```
